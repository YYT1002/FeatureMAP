Read the expression data and pseudotime.

```python
import pandas as pd
# LI data
data = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-LI/dyn-LI-5000-1/ExpressionData.csv', index_col=0).T
data_pseudotime = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-LI/dyn-LI-5000-1/Pseudotime.csv', index_col=0)
# replace Nan with 0
data_pseudotime.fillna(0, inplace=True)
# sum PseudoTime1 and PseudoTime2 
data_pseudotime = data_pseudotime.sum(axis=1)


# Cycle data
data = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-CY/dyn-CY-5000-1/ExpressionData.csv', index_col=0).T
data_pseudotime = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-CY/dyn-CY-5000-1/Pseudotime.csv', index_col=0)
# replace Nan with 0
data_pseudotime.fillna(0, inplace=True)
# sum PseudoTime1 and PseudoTime2 
data_pseudotime = data_pseudotime.sum(axis=1)



# BF data
data = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-BF/dyn-BF-5000-1/ExpressionData.csv', index_col=0).T
data_pseudotime = pd.read_csv('./BEELINE-data/inputs/Synthetic/dyn-BF/dyn-BF-5000-1/Pseudotime.csv', index_col=0)
# replace Nan with 0
data_pseudotime.fillna(0, inplace=True)
# sum PseudoTime1 and PseudoTime2 
data_pseudotime = data_pseudotime.sum(axis=1)




# # BFC data
# data = pd.read_csv('./data/BEELINE-data/inputs/Synthetic/dyn-BFC/dyn-BFC-5000-1/ExpressionData.csv', index_col=0).T
# data_pseudotime = pd.read_csv('./data/BEELINE-data/inputs/Synthetic/dyn-BFC/dyn-BFC-5000-1/Pseudotime.csv', index_col=0)
# # replace Nan with 0
# data_pseudotime.fillna(0, inplace=True)
# # sum PseudoTime1 and PseudoTime2 
# data_pseudotime = data_pseudotime.sum(axis=1)

# # TF data
# data = pd.read_csv('./data/BEELINE-data/inputs/Synthetic/dyn-TF/dyn-TF-5000-1/ExpressionData.csv', index_col=0).T
# data_pseudotime = pd.read_csv('./data/BEELINE-data/inputs/Synthetic/dyn-TF/dyn-TF-5000-1/Pseudotime.csv', index_col=0)
# # replace Nan with 0
# data_pseudotime.fillna(0, inplace=True)
# # sum PseudoTime1 and PseudoTime2 
# data_pseudotime = data_pseudotime.sum(axis=1)



# # curated data
# data = pd.read_csv('./data/BEELINE-data/inputs/Curated/mCAD/mCAD-2000-1/ExpressionData.csv', index_col=0).T
# data_pseudotime = pd.read_csv('./data/BEELINE-data/inputs/Curated/mCAD/mCAD-2000-1/Pseudotime.csv', index_col=0)
# # replace Nan with 0
# data_pseudotime.fillna(0, inplace=True)
# # sum PseudoTime1 and PseudoTime2 
# data_pseudotime = data_pseudotime.sum(axis=1)
```
